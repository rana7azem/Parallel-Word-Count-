```python
import matplotlib.pyplot as plt

threads = [1,2,4,8]
speedup = [1.0,1.402,2.574,3.271]
efficiency = [1.0,0.701,0.643,0.409]


plt.figure()
plt.plot(threads, speedup, marker='o')
plt.xlabel('Threads (p)')
plt.ylabel('Speedup (T_S/T_P)')
plt.title('Speedup vs Threads')
plt.grid(True)
plt.show()


plt.figure()
plt.plot(threads, efficiency, marker='o')
plt.xlabel('Threads (p)')
plt.ylabel('Efficiency (Speedup/p)')
plt.title('Efficiency vs Threads')
plt.grid(True)
plt.show()
```
